```python
from sklearn.metrics import r2_score 
R_square_250 = r2_score(y_train_Sensitivity_250, predicted_on_X_train_Sensitivity_250)
print('R_Square_250', R_square_250) 
R_square_1000 = r2_score(y_train_Sensitivity_1000, predicted_on_X_train_Sensitivity_1000)
print('R_Square_1000', R_square_1000) 
R_square_5000 = r2_score(y_train_Sensitivity_5000, predicted_on_X_train_Sensitivity_5000)
print('R_Square_5000', R_square_5000) 
R_square_10000 = r2_score(y_train_Sensitivity_10000, predicted_on_X_train_Sensitivity_10000)
print('R_Square_10000', R_square_10000) 
```

```python
plt.figure(figsize = (7,7))
sns.set_style("whitegrid") 

xx= y_train_Sensitivity_5000 
yy = predicted_on_X_train_Sensitivity_5000
bubble_plot_line_x1y1 = [min(np.minimum(xx,yy)), max(np.maximum(xx,yy))]
bubble_plot_line_x2y2 = [min(np.minimum(xx,yy)), max(np.maximum(xx,yy))]
plt.xlim(bubble_plot_line_x1y1[0], bubble_plot_line_x1y1[1])
plt.ylim(bubble_plot_line_x1y1[0], bubble_plot_line_x1y1[1])

plt.plot(bubble_plot_line_x1y1, bubble_plot_line_x2y2, 'k-', linewidth=2)

plt.plot(y_train_Sensitivity_250, predicted_on_X_train_Sensitivity_250, 'm*', markersize = 15, markeredgewidth = 3,
         markerfacecolor = 'None', label = 'Predicted Sensitivity - Epoch 250')

plt.text(50, 450, 'R_Squared_250 = %0.4f' % R_square_250, fontsize = 12, fontweight = 'bold')

plt.plot(y_train_Sensitivity_1000, predicted_on_X_train_Sensitivity_1000, 'b*', markersize = 15, markeredgewidth = 3,
         markerfacecolor = 'None', label = 'Predicted Sensitivity - Epoch 1000')

plt.text(50, 430, 'R_Squared_1000 = %0.4f' % R_square_1000, fontsize = 12, fontweight = 'bold')

plt.plot(y_train_Sensitivity_5000, predicted_on_X_train_Sensitivity_5000, 'g*', markersize = 15, markeredgewidth = 3,
         markerfacecolor = 'None', label = 'Predicted Sensitivity - Epoch 5000')

plt.text(50, 410, 'R_Squared_5000 = %0.4f' % R_square_5000, fontsize = 12, fontweight = 'bold')

#plt.plot(y_validation_Sensitivity_5000, predicted_on_X_validation_Sensitivity_5000, 'ko', markersize = 12, 
        # markeredgewidth = 4, markerfacecolor = 'None', label = 'Actual Sensitivity')

plt.plot(y_train_Sensitivity_10000, predicted_on_X_train_Sensitivity_10000, 'r*', markersize = 15, markeredgewidth = 3,
         markerfacecolor = 'None', label = 'Predicted Sensitivity - Epoch 10000')

plt.text(50, 390, 'R_Squared_10000 = %0.4f' % R_square_10000, fontsize = 12, fontweight = 'bold')

legend_properties = {'weight':'bold'}

plt.legend(loc = 'best', fontsize = 15, prop = legend_properties)
plt.tick_params(axis = "both", labelsize = 15, width = 58769)

plt.xlabel('Actual Sensitivity (nm/RIU)', fontsize = 18, fontweight = 'bold')
plt.ylabel('Predicted Sensitivity (nm/RIU)', fontsize = 18 , fontweight = 'bold')
plt.title('Actual vs Predicted Sensitivity 4(h)', fontsize = 20 , fontweight = 'bold', color = 'black')

fontsize = 15
ax = gca()
for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('bold')
    
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('bold')
```
